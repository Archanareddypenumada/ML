```python
#Importing requird datasets
import pandas as pd
import numpy as np 
#Converting the dataset into pandas dataframe
df=pd.read_csv("C:/Users/year3/Desktop/purchase_data.csv")
#Creating A and B dataframes with the features and the resultant columns respectively
A=df[['Candies (#)','Mangoes (Kg)','Milk Packets (#)']]
C=df[['Payment (Rs)']]
#Finding the dimensionality, number of vectors, rank of the data A
dim=A.shape[1]
num_vectors=A.shape[0]
rank_A=np.linalg.matrix_rank(A)
#Finding the pseudo inverse of A and finding thhe dot product of A pseudo inverse and C to find the X matrix
pseudo_inv_A=np.linalg.pinv(A)
X=pseudo_inv_A.dot(C)
print("Dimensionality of the data is ",dim)
print("Number of vectors in the vector space is ",num_vectors)
print("Costs of each product are: ",X)
#Categorizing the customers into 2 classes:rich and poor based on the bill
df['Class']=df['Payment (Rs)'].apply(lambda y: 'Rich' if y>200 else 'Poor')
print(df)
```

```python
#Importing requird datasets
import pandas as pd
import numpy as np
#Converting the dataset into pandas dataframe
df1=pd.read_csv("C:/Users/year3/Desktop/irctc.csv")
#Cleaning the data by removing the commas from the prices
cleaned_data=pd.to_numeric(df1['Price'].str.replace(',',''))
#Finding the mean and variance of the cleaned data
mean=np.mean(cleaned_data)
print("Mean of the data is",mean)
variance=np.var(cleaned_data)
print("Variance of the data is",variance)
#Creating a dataframe with data where day is wednesday
wed_data=df1[df1['Day']=='Wed']
#Cleaning the data by removing the commas from the prices
cleaned_wed_data=pd.to_numeric(wed_data['Price'].str.replace(',',''))
#Finding the mean of the cleaned data
mean_wed=np.mean(cleaned_wed_data)
print("Mean of the price on wednesday is",mean_wed)
# Selecting the price data for the month of April
df1['Date']=pd.to_datetime(df1['Date'])
april_data = df1[df1['Date'].dt.month == 4]
mean_april_price = np.mean(april_data['Price'])
print("Mean Price in April:", mean_april_price)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv("C:/Users/year3/Desktop/irctc1.csv")

# Replace 'column_name' with the actual name of your columns
price_column_name = 'Price'
day_column_name = 'Day'
month_column_name = 'Month'
chg_column_name = 'Chg%'

# Extract the columns
price_data = data[price_column_name]
day_data = data[day_column_name]
month_data = data[month_column_name]
chg_data = data[chg_column_name]

# Remove commas and convert the price column to numeric (errors='coerce' will convert problematic entries to NaN)
price_data_cleaned = pd.to_numeric(price_data.str.replace(',', ''), errors='coerce')

# Drop any NaN values that might have been introduced
price_data_cleaned = price_data_cleaned.dropna()

# Filter data for Wednesdays
wednesday_data = data[day_data == 'Wed']
wednesday_price_data = pd.to_numeric(wednesday_data[price_column_name].str.replace(',', ''), errors='coerce')
wednesday_price_data = wednesday_price_data.dropna()

# Filter data for April
april_data = data[month_data == 'Apr']
april_price_data = pd.to_numeric(april_data[price_column_name].str.replace(',', ''), errors='coerce')
april_price_data = april_price_data.dropna()

# Calculate the mean and variance for all data
mean_all = price_data_cleaned.mean()
variance_all = price_data_cleaned.var()

# Calculate the mean price for Wednesdays
mean_wednesday = wednesday_price_data.mean()

# Calculate the mean price for April
mean_april = april_price_data.mean()

# Print the results
print(f"Mean of {price_column_name}: {mean_all}")
print(f"Variance of {price_column_name}: {variance_all}")
print(f"Mean price on Wednesdays: {mean_wednesday}")
print(f"Mean price in April: {mean_april}")

# Compare mean prices
print(f"\nComparison of Mean Prices:")
print(f"Mean price on Wednesdays is {'higher' if mean_wednesday > mean_all else 'lower' if mean_wednesday < mean_all else 'the same as'} the overall mean price.")
print(f"Mean price in April is {'higher' if mean_april > mean_all else 'lower' if mean_april < mean_all else 'the same as'} the overall mean price.")

# Clean the Chg% column
chg_data_cleaned = pd.to_numeric(chg_data.str.replace('%', '').str.replace(',', ''), errors='coerce')
chg_data_cleaned = chg_data_cleaned.dropna()

# Calculate the probability of making a loss (where Chg% is less than 0)
probability_of_loss = (chg_data_cleaned < 0).mean()

print(f"Probability of making a loss: {probability_of_loss:.2%}")

# Extract and clean the Chg% column for Wednesdays
chg_data_wednesday = wednesday_data[chg_column_name]
chg_data_wednesday_cleaned = pd.to_numeric(chg_data_wednesday.str.replace('%', '').str.replace(',', ''), errors='coerce')
chg_data_wednesday_cleaned = chg_data_wednesday_cleaned.dropna()

# Calculate the probability of making a profit on Wednesdays (where Chg% is greater than 0)
probability_of_profit_wednesday = (chg_data_wednesday_cleaned > 0).mean()

print(f"Probability of making a profit on Wednesdays: {probability_of_profit_wednesday:.2%}")

# Map days of the week to numeric values for plotting
day_mapping = {
    'Mon': 0,
    'Tue': 1,
    'Wed': 2,
    'Thu': 3,
    'Fri': 4,
    'Sat': 5,
    'Sun': 6
}
day_numeric = day_data.map(day_mapping)

# Clean the Chg% column for plotting
chg_data_cleaned = pd.to_numeric(chg_data.str.replace('%', '').str.replace(',', ''), errors='coerce')
chg_data_cleaned = chg_data_cleaned.dropna()
day_numeric = day_data[chg_data_cleaned.index].map(day_mapping)

# Create a DataFrame for plotting
plot_data = pd.DataFrame({
    'Day': day_numeric,
    'Chg%': chg_data_cleaned
})

# Set up the plotting environment
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Day', y='Chg%', data=plot_data)

# Label the axes and add a title
plt.xticks(ticks=list(day_mapping.values()), labels=list(day_mapping.keys()))
plt.xlabel('Day of the Week')
plt.ylabel('Chg%')
plt.title('Scatter Plot of Chg% Data Against Day of the Week')

# Show the plot
plt.show()
```
